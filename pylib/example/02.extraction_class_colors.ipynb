```python
from pylib.util import retrieve_n_class_color_cubic, retrieve_n_class_color_sphere
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def plot_n_class_colors(Nclass):
    class_color_1 = retrieve_n_class_color_cubic(Nclass)
    class_color_2 = retrieve_n_class_color_sphere(Nclass)

    x = range(Nclass)
    y = [1 for i in x]

    fig = plt.figure(figsize=[10, 4])

    ax1 = fig.add_subplot(2,1,1)
    objs = ax1.bar(x, y)

    for i in range(Nclass):
        objs[i].set_color(class_color_1[i])
    ax1.set_xlim([-0.2, Nclass + 0.2])
    ax1.set_ylim([-0.1, 1.1])
    ax1.set_xlabel('class #')
    ax1.set_title('Cubic representation')

    ax2 = fig.add_subplot(2,1,2)
    objs = ax2.bar(x, y)

    for i in range(Nclass):
        objs[i].set_color(class_color_2[i])
    ax2.set_xlim([-0.2, Nclass + 0.2])
    ax2.set_ylim([-0.1, 1.1])
    ax2.set_xlabel('class #')
    ax2.set_title('Sphere representation')

    fig.tight_layout()

    fig.show()
```

```python
plot_n_class_colors(20)
```

```python
plot_n_class_colors(100)
```
